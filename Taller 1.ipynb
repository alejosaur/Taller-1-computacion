**Solución**


Sea M=(Q,$\Sigma$,$\delta$,q0,$F_{1}$)

entonces, definimos un autómata M'en el cual, los estados finales $F_{2}$ se definan de la siguiente manera: Los estados pertenecientes a $F_{1}$ no sean estados finales de M', y los estados pertenecientes a q, pero que no pertenecen a $F_{1}$, sean los estados finales.]

M'=(Q,$\Sigma$,$\delta$,q0,$F_{2}$)

$F_{2}=\displaystyle\bigcup_{k=1}^{n}{(q_{n} \in Q : q_{n} \notin F_{1})}$

**Cuarto punto**  
Dados dos autómatas finitos determinísticos $M_{1}$ y $M_{2}$ con el mismo alfabeto de entrada, construya un tercer autómata finito determinístico ($M$) con la propiedad que $u \in \Sigma$* es aceptado por $M$ si y solo si es aceptado por ambos $M_{1}$ y $M_{2}$. 

**Solución**

Sea    
$M_{1}=(Q_{1},\Sigma,\delta_{1},q_{0_{1}},F_{1})$  

y  

$M_{2}=(Q_{2},\Sigma,\delta_{2},q_{0_{2}},F_{2})$  

entonces procedemos a definir el autómata $M$:  

$M=((Q_{1}XQ_{2}),\Sigma,(\delta_{1},\delta_{2}),(q_{0_{1}},q_{0_{2}}),(F_{1}XF_{2}))$  

En el cual definimos en conjunto de estados mediante el producto cruz entre $\{q_{\{1}$ y $\{q_{\{2}$, lo cual nos da todas las parejas posibles entre los estados de $M_{1}$ y $M_{2}$; El lenguaje de entrada es el mismo, la función de transición está dada por (\delta_{1},\delta_{2}), la cual nos lleva de un estado a otro en M, ya que estos están compuestos por "parejas"; El estado inicial es el estado compuesto por la pareja de estados iniciales de $\{q_{\{1}$ y $\{q_{\{2}$; y el estado final es el producto entre los estados finales $F_{1}$ y los estados finales $F_{2}$ , lo cual nos da como resultado todas las parejas de estados finales.
